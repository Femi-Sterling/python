**1A. Python Calculator**

```
a,b = eval(input("Enter two numbers: "))
print("1.Addition\t2.Subtraction\t3.Multiplication\t4.Division\t5.Modulus")
ch = int(input("Enter your choice: "))
if ch == 1:
  print("Sum: ",a+b)
elif ch == 2:
  print("Difference: ",a-b)
elif ch == 3:
  print("Product: ",a*b)
elif ch == 4:
  print("Division: ",a/b)
elif ch == 5:
  print("Modulus: ",a%b)
else:
  print("Invalid choice.")
```
